```python
import networkx as nx
from reducer import*
import matplotlib.pyplot as plt
```

```python
def show_graph(graph):
    # function for plotting a graph (directed or undirected)
    pos = nx.spring_layout(graph)
    nx.draw(graph, pos, node_size=500, with_labels=True, arrows='directed', connectionstyle='arc3, rad = 0.1')
    plt.show()
    

g = nx.DiGraph()
g.add_edge('u', 't')
g.add_edge('t', 'm')
g.add_edge('t','y')
g.add_edge('t', 'q')
g.add_edge('y', 'q')
g.add_edge('m', 'n')
g.add_edge('n', 'y')
g.add_edge('c', 'm')
g.add_edge('c', 'n')
g.add_edge('c', 't')
g.add_edge('u', 'n')
g.add_edge('c', 'k')
g.add_edge('k', 'y')
g.add_edge('k', 'r')
g.add_edge('r', 'y')
g.add_edge('q', 'p')
g.add_edge('u', 'y')
g.add_edge('z', 't')
g.add_edge('x', 'y')
xs = ['t']
ys = ['y']


g1 = nx.DiGraph()
g1.add_edge('att2', 'att3')
g1.add_edge('att3', 'att4')
g1.add_edge('l2','l3')
g1.add_edge('l3', 'l4')
g1.add_edge('l3', 'att4')
g1.add_edge('att4', 'l4')
g1.add_edge('c1', 'att2')
g1.add_edge('c1', 'l2')
g1.add_edge('c1', 'att3')
g1.add_edge('c2', 'l2')
g1.add_edge('c2', 'l3')
g1.add_edge('c2', 'l4')
g1.add_edge('c1', 'att4')
g1.add_edge('c2', 'att3')

xs1 = ['att2', 'l2']
ys1 = ['att4', 'l4']

print('Full Graph:')
show_graph(graph=g)
rd = reducer(g,xs,ys, precision_vars=False)
print('Reduced Graph:')
show_graph(graph=rd)

print('Full Graph:')
show_graph(graph=g1)
rd1 = reducer(g1,xs1,ys1, precision_vars=False)
print('Reduced Graph:')
show_graph(graph=rd1)
```
